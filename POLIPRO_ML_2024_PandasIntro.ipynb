A forma que filtramos dados no Pandas é similar a expressões booleanas do Python. A diferença é que ao invés de usarmos and (&&) e or (||), usamos & e |. Por exemplo, considere selecionar itens que tenham a coluna F >= 1.2 E que a coluna E seja do tipo "train". Teríamos:

```python
df[(df["F"] >= 1.2) & (df["E"] == "train")]
```

Perceba que selecionamos cada coluna (ex: df["F"]) e escrevemos a expressão dentro de parênteses. Cada expressão deve estar dentro de um parenteses.

Para o mesmo caso, mas imagine que necessitamos de retornar itens que a coluna F seja maior ou igual a 1.2 OU que a coluna E seja do tipo "train":

```python
df[(df["F"] >= 1.2) | (df["E"] == "train")]
```


Perceba que neste caso, ao invés de duas linhas são retornadas 3 linhas. Pois item 2 não é "train", mas seu valor de F é maior ou igual a 1.2


Caso tenhamos uma lista de elementos para procurar no DataFrame. Também podemos usar o operador isin(). Por exemplo, imagine que gostaríamos de retornar linhas que tenham como valor de C=1 ou C=3. Poderíamos fazer isto de duas formas:

```python
df[(df["C"] == 1) | (df["C"] == 3)]
```

```python
df[df["C"].isin([1,3])]
```

Assim como no SQL, podemos agrupar elementos no pandas. Para isso usamos o método df.groupby([coluna1, coluna2, ...]). A partir disso podemos usar funcoes como COUNT (para contar a quantidade de elementos). Isso é feito da seguinte forma:

```python
df.groupby(['E'])['E'].agg(['count'])
```

Ou diretamente no metodo:

```python
df.groupby(['E'])['E'].count()
```

Também é possivel usar as funcoes median, mean, std caso estejamos trabalhando com variaveis numericas

As vezes nossos dados podem estar em dois DataFrames distintos, mas que podem ser juntados assim como um SQL JOIN. Considere o seguintes DataFrames:

```python
df2 = pd.DataFrame(
{
   "C": pd.Series(list(range(3))),
   "G": [5, 5, 33],
   "L": [None, "total", None],
 }
)
df2
```

```python
df = pd.DataFrame(
{
   "A": 1.0,
   "B": pd.Timestamp("20130102"),
   "C": pd.Series(list(range(4))),
   "D": [3] * 4,
   "E": pd.Categorical(["test", "train", "test", "train"]),
   "F": [None, 1.4, 1.2, 3],
 }
)
df
```

Podemos juntá-los usando o método merge usando left (similar a um LEFT JOIN em SQL):

```python
df.merge(df2, on="C", how='left')
```

Perceba que a linha 3 não possui valores para G e L pois df2 não possuía C=3

Podemos juntá-los usando o método merge usando right (similar a um INNER JOIN em SQL):

```python
df.merge(df2, on="C", how='inner')
```

Perceba que agora a linha 3 não é apresentada

Muitas vezes precisaremos de realizar transformações em dados. Um método que usaremos constantemente nessa tarefa é o método apply. Primeiramente, considere a seguinte função de transformação:

```python
def fn(x):
  if x > 1.2:
    return "Valido"
  else:
    return "Invalido"
```

Vamos agora verificar usando a funcao fn quais valores de df['F'] sao validos. Para isso utilizaremos a função apply:

```python
df['F2']=df['F'].apply(fn)
df
```

Perceba que agora temos a coluna F2 no Dataframe.

Existe uma forma mais elegante de se fazer isso! Não usaremos a definição da função fn, mas uma função lambda. Uma funcao anonima em Python, ou seje, que não precisa ser declarada.

```python
df['F2']=df['F'].apply(lambda x : "Valido" if x > 1.2 else "Invalido")
df
```
